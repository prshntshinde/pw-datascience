## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

- Scatter plots are the graphs that present the relationship between two variables in a data-set. 
- It represents data points on a two-dimensional plane. 
- The independent variable or attribute is plotted on the X-axis, while the dependent variable is plotted on the Y-axis.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y, c='#d62728')
plt.title('Scatter Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

- Subplot function is used add an Axes to the current figure or retrieve an existing Axes.

```python
# Plot 1
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y, c='r')
plt.title('Plot 1')

# Plot 2
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y, c='g')
plt.title('Plot 2')

# Plot 3
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y, c='b')
plt.title('Plot 3')

# Plot 4
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y, c='#F60DB6')
plt.title('Plot 4')

plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

- Bar plots are a type of data visualization used to represent data in the form of rectangular bars. 
- The height of each bar represents the value of a data point, and the width of each bar represents the category of the data.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
```

```python
plt.barh(company,profit)
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

- Box plot displays the five-number summary of a set of data. 
- The five-number summary is the minimum, first quartile, median, third quartile, and maximum. 
- In a box plot, we draw a box from the first quartile to the third quartile. 
- A vertical line goes through the box at the median.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot([box1, box2])
```
