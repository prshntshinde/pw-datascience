```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```

```python
plt.scatter(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("X vs Y")
```

```python
plt.scatter(x,y, c='red')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("X vs Y")
```

```python
plt.scatter(x,y, c='red', alpha=.6)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("X vs Y")
```

```python
plt.scatter(x,y, c='red', alpha=.6, marker='X')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("X vs Y")
```

```python
plt.scatter(x,y, c='red', alpha=.6, marker='X')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("X vs Y")
plt.grid()
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y, c='red', alpha=.6, marker='X')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("X vs Y")
plt.grid()
```
